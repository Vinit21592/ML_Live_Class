```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\TERMINATOR\Desktop\ML Example\Modular\ML_Live_Class\data\Advertising.csv")
df.head()
```

```python
df['total_spent'] = df['TV'] + df['radio'] + df['newspaper']
sns.scatterplot(data=df , x='total_spent', y='sales')
```

```python
sns.regplot(data=df, x='total_spent', y='sales', color='g') #seaborn is able to draw a regression line using OLS
```

```python
x = df['total_spent']
y = df['sales']
```

```python
np.polyfit(x,y, deg=1) #it's a polynomial fit based on non linear least squares or ordinary least squares
#This function uses least squares and the solution is to minimize the squared errors in the given polynomial.
```

```python
#Numpy.polyfit and Scikit-learn.LinearRegression with PolynomialFeatures might be used interchangeably because those return the same coefficients and interce
```
